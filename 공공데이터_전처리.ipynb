# 2023 1월~10월까지

```python
import pandas as pd


combined_df2 = pd.DataFrame()


sigungu_codes = ["11110", "11140", "11170",'11200','11215','11230','11260','11290','11305','11320','11350','11380','11380','11410','11440','11470'
,'11500','11530','11545','11560','11590','11620','11650','11680','11710','11740']
for sigungu_code in sigungu_codes:
    df = api.get_data(
        property_type="아파트",
        trade_type="매매",
        sigungu_code=sigungu_code,
        start_year_month="202301",
        end_year_month="202310",
    )

    # 가져온 데이터프레임을 기존의 combined_df에 추가
    combined_df2 = pd.concat([combined_df2, df], ignore_index=True)
```

#면적당 가격 열 추가

```python
combined_df2['전용면적당 가격']=combined_df2['거래금액'] / combined_df2['전용면적']
```

```python
sigungu_mapping = {
    "11110": "종로구",
    "11140": "중구",
    "11170": "용산구",
    "11200": "성동구",
    "11215": "광진구",
    "11230": "동대문구",
    "11260": "중랑구",
    "11290": "성북구",
    "11305": "강북구",
    "11320": "도봉구",
    "11350": "노원구",
    "11380": "은평구",
    "11410": "서대문구",
    "11440": "마포구",
    "11470": "양천구",
    "11500": "강서구",
    "11530": "구로구",
    "11545": "금천구",
    "11560": "영등포구",
    "11590": "동작구",
    "11620": "관악구",
    "11650": "서초구",
    "11680": "강남구",
    "11710": "송파구",
    "11740": "강동구"
}

# 자치구 정보를 추가
combined_df2['자치구'] = combined_df2['지역코드'].map(sigungu_mapping)
```

```python
combined_df2 = combined_df2[['자치구'] + [col for col in combined_df2.columns if col != '자치구']]
```

```python
combined_df2.to_csv('2023아파트매매가.csv', index=False, encoding='utf-8')
```

```python
df=combined_df2[['자치구','전용면적당 가격']]
```

```python
result_df = df.groupby('자치구')['전용면적당 가격'].mean().reset_index()
result_df.rename(columns={'거래금액': '면적당평균거래금액'}, inplace=True)

print(result_df)
```

```python
result_df.to_csv('2023_자치구별_면적당_평균_아파트매매가.csv', index = False, encoding='utf-8-sig')
```
